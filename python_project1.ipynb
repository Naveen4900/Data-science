## Tesla revenue data by using webscraping

```python
!pip install bs4
!pip install requests
!pip install pandas

from bs4 import BeautifulSoup
import requests 
import pandas as pd

url = "https://finance.yahoo.com/quote/TSLA/financials/"

data = requests.get(url).text
print(data)

soup = BeautifulSoup(data,"html.parser")

revenue_data = soup.find('div', class_='Total Revenue')
print(revenue_data)
```

## make a graph on tesla_stock

```python
!pip install yfinance
!pip install matplotlib
import yfinance as yf
import matplotlib.pyplot as plt

# Function to plot the stock data
def make_graph(stock_data, title="Stock Data"):
    """
    This function takes stock data and creates a plot.
    
    :param stock_data: DataFrame containing stock data.
    :param title: Title of the graph.
    """
    # Plotting the stock's closing price
    stock_data['Close'].plot(title=title, figsize=(10, 6))
    
    # Set labels and title
    plt.xlabel("Date")
    plt.ylabel("Close Price ($)")
    plt.title(title)
    
    # Show the plot
    plt.show()

# Download Tesla stock data for the past 1 year
tesla_data = yf.download('TSLA', period='1y', progress=False)

# Call the make_graph function to plot Tesla stock data
make_graph(tesla_data, title="Tesla Stock Data (1 Year)")
```
